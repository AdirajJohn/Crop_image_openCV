### Croping the image using openCV

#### Import Necessary Library

```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import os
```

#### Load the image

```python
image=cv2.imread("D:/repository/Crop_image_openCV/TO_Crop/W3-1002_20230815_100939.png")
```

```python
print(image.shape)
```

#### Display the Image

```python
cv2.imshow("sample",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
croped_image=image[0:325,0:363]
```

```python
plt.imshow(croped_image)
```

#### Save the Image

```python
cv2.imwrite("D:/repository/Crop_image_openCV/After_crop/W3-1002_20230815_100939.png",croped_image)
```
